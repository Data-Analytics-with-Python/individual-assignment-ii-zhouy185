```python
product_cats = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

df_melted = df.melt(id_vars=['Education'], value_vars=product_cats, var_name='ProductCategory', value_name='Spending')

fig = px.bar(df_melted,
             x='Education',
             y='Spending',
             color='ProductCategory',
             title='Spending by Education Level and Product Category',
             labels={'Education': 'Education Level', 'Spending': 'Total Spending', 'ProductCategory': 'Product Category'}) # Use 'relative' for stacked bars where the total height shows combined spending

fig.update_layout(xaxis_title='Education Level', yaxis_title='Total Spending')
fig.update_traces(opacity=1, marker_line_width=0) # Remove border lines as per hint

fig.show()
```

```python
# First label the generation for each customer

def year_to_gen(year):
  if (year >=1981) &  (year <= 1996):
    return 'Millennials'
  if  (year >=1965) &  (year <= 1980):
    return 'Gen. X'
  else:
    return 'Other'

df['Generation'] = df['Year_Birth'].apply(year_to_gen)

df = df[df['Generation']!='Other']

df.head()
```

```python
fig = px.box(df, x='Generation',y='NumWebPurchases',width=500,height=400)

fig.update_layout(yaxis_title='Number of Online Purcahses', xaxis_title='Generation')

fig.show()
```

```python
fig = px.box(df, x='Generation',y='NumWebPurchases',color='Kidhome',width=500,height=400)

fig.update_layout(yaxis_title='Number of Online Purcahses', xaxis_title='Generation')

fig.show()
```
